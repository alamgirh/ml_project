```
import matplotlib.pyplot as plt
import numpy as np
X = ['3DDFA\n(BIWI)','Multi-Loss AlexNet\n(AFLW2000)','Hopenet\n(300W-LP)','Hopenet\n(SFU-Store-Nav)']
Yaw = [36.175,27.650,6.8725,38.6746]
Pitch = [12.252,8.543,6.64565,14.2506]
Roll = [8.776,8.954,5.68285,15.3274]
#MAE = [19.068, 15.049, 6.410, 35.0842]
MAE = [(Yaw[0]+Pitch[0]+Roll[0])/3, (Yaw[1]+Pitch[1]+Roll[1])/3, (Yaw[2]+Pitch[2]+Roll[2])/3, (Yaw[3]+Pitch[3]+Roll[3])/3]

_X = 2*np.arange(len(X))

plt.bar(_X - 0.2, Yaw, 0.4, label='Yaw')
plt.bar(_X + 0.2, Pitch, 0.4, label='Pitch')
plt.bar(_X + 0.6, Roll, 0.4, label='Roll')
plt.bar(_X + 1.0, MAE, 0.4, label='MAE')
plt.ylabel('Error in degree')
plt.xticks(_X, X) # set labels manually
plt.legend()
plt.grid(True)

plt.show()
```
